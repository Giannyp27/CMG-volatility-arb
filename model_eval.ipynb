```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/gianlucapannozzo/Desktop/university/VS Code/dltr-volatility-arb/data/vol_predictions.csv", parse_dates=["timestamp"])

plt.figure(figsize=(14, 6))
plt.plot(df["timestamp"], df["actual_vol"], label="Actual Volatility", linewidth=2)
plt.plot(df["timestamp"], df["predicted_vol"], label="Predicted Volatility (LSTM)", linestyle='--')
plt.title("LSTM Volatility Forecast vs Actual")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(df["actual_vol"], df["predicted_vol"])
rmse = np.sqrt(mean_squared_error(df["actual_vol"], df["predicted_vol"]))
r2 = r2_score(df["actual_vol"], df["predicted_vol"])

print(f"MAE:  {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R^2:  {r2:.4f}")
```
